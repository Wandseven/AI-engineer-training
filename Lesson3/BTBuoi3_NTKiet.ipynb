## Bài tập buổi 3

### Bài 1: Tính số ngày trong tháng

```python
month = int(input('Nhập vào tháng cần kiểm tra: '))

if month == 2:
    leap_year = input('Có phải năm nhuận không? (y/n): ')
    day = 29 if leap_year == 'y' else 28
elif month == 4 or month == 6 or month == 9 or month == 11:
    day = 30
else:
    day = 31


if 1 <= month <= 12: print(f'Tháng {month} có {day} ngày.')
else: print('Tháng không hợp lệ.')
```

### Bài 2: Kiểm tra năm nhuận

```python
year = int(input('Nhập vào năm cần kiểm tra: '))

if year % 4 == 0 and year % 100 != 0:
    leap_year = True
elif year % 400 == 0:
    leap_year = True
else:
    leap_year = False

if year > 0:
    if leap_year == True:
        print(f'Năm {year} là năm nhuận.')
    else:
        print(f'Năm {year} không phải năm nhuận.')
else:
    print('Năm không hợp lê.')
```

### Bài 3: Tính tiền điện

```python
k = int(input('Nhập vào số điện (kWh): '))

if 0 <= k <= 50:
    price = k*1.678
elif 50 < k <= 100:
    price = 50*1.678 + (k-50)*1.734
elif 100 < k <= 200:
    price = 50*1.678 + 50*1.734 + (k-100)*2.014
else:
    price = 50*1.678 + 50*1.734 + 100*2.014 + (k-200)*2.536

if k >= 0:
    print(f'Tiền điện là {price:.2f} VND')
else:
    print('Số điện không hợp lệ.')
```

### Bài 4: Xếp loại sức khỏe

```python
bmi = float(input('Nhập vào chỉ số BMI: '))

if 0 < bmi < 18.5:
    body_status = 'Gầy'
elif 18.5 <= bmi < 25:
    body_status = 'Bình thường'
elif 25 <= bmi < 30:
    body_status = 'Thừa cân'
else:
    body_status = 'Béo phì'

if bmi > 0:
    print(f'Xếp loại sức khỏe của bạn là: {body_status}.')
else:
    print('Chỉ số BMI không hợp lệ.')
```

### Bài 5: Tính tổng bình phương các số

```python
N = int(input('Nhập N: '))
sum = 0

for i in range(N):
    sum += (i+1)**2

if N > 0: print(f'Tổng bình phương các số từ 1 đến {N} là: {sum}')
else: print('N không hợp lệ.')
```

### Bài 6: Nhập danh sách và tính tổng

```python
n = 5
sum = 0

for i in range(n):
    sum += int(input(f'Nhập vào số thứ {i+1}: '))

print(f'Tổng 5 số vừa nhập là: {sum}')
```

### Bài 7: Tính tổng các số nhập vào

```python
positive = True
sum = 0

while positive:
    a = int(input('Nhập vào số'))
    if a >= 0: sum += a
    else: positive = False
    
print(f'Tổng các số dương là: {sum}')
```

### Bài 8: Số Fibonacci

```python
f0 = 0
f1 = 1
f2 = 0

print(f'{f0} {f1}', end=' ')

while f2 < 100:
    f2 = f1 + f0
    f0 = f1
    f1 = f2
    if f2 < 100: print(f'{f2}', end=' ')
```

### Bài 9: Tính tổng các số chẵn từ 1 đến N:

```python
N = int(input('Nhập vào N: '))
sum = 0

for i in range(N):
    if (i+1) % 2 == 0:
        sum += (i+1)

print(f'Tổng các số chẵn từ 1 đến {N} là: {sum}')
```

### Bài 10: Tính trung bình cộng

```python
positive = True
sum = 0
N = 0

while positive:
    a = int(input('Nhập vào số'))
    if a > 0:
        N += 1
        sum += a
    else:
        positive = False

aver = sum/N 
print(f'Trung bình cộng các số vừa nhập là: {aver}')
```
