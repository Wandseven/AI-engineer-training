```python
s = 0
n = int(input('Enter n: '))
for i in range(1, n + 1):
    s += i      # s = s + i
print(f'Sum 1-{n} = {s}')
```

```python
n = int(input('Enter n: '))
s = 0
for _ in range(n):
    s += int(input('Enter your number: '))
print(f'Sum of {n} numbers = {s}')
```

```python
n = int(input('Enter n: '))
count = 0
for _ in range(n):
    a = int(input('Enter a number: '))
    if a & 2 == 0:
        count += 1
print(f'Count of even number = {count}')
```

```python
n = int(input('Enter n: '))
for i in range (1, 11):
    print(f'{i:2} x {n} = {i*n:2}')
```

```python
n = int(input('Enter n: '))
for i in range(n):
    for j in range(i+1):
        print('*', end=' ')
    print()
```
