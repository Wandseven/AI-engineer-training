```python
# Nhập 1 ngày trong tuần và in ra là ngày làm việc hay ngày cuối tuần
day = int(input('Enter a day in the week: '))
if day < 1 or day > 7:
    status = 'Invalid day'
elif day == 1 or day == 7:
    status = 'Weekend'
else:
    status = 'Working day'
print(f'{status}')
```

```python
payment = int(input('Enter your payment: '))
member_status = input('Are you a member (Y/N): ')

if member_status == 'Y' or member_status == 'y':
    membership = input('Enter membership type (VIP/Gold/Diamond): ')
    if membership == 'VIP': payment *= 0.9
    elif membership == 'Gold': payment *= 0.85
    elif membership == 'Diamond': payment *= 0.75
    else: print('Invalid membership type')

print(f'Final payment amount: {payment:.2f} VND')
```
