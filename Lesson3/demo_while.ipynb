```python
printing = True
while printing:
    n = int(input('Enter n: '))
    for i in range(1,11):
        print(f'{i:2} x {n} = {i * n:2}')
    
    answer = input('Do you want to continue? (yes/no): ')
    printing = answer == 'yes'
print('Goodbye!')
```
