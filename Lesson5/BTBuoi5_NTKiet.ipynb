### Bài 1: Nhân đôi các phần tử trong List

```python
a = rd.sample(range(100), 10)
print(a)

for i in range(len(a)):
    a[i] = a[i] * 2

print(a)
```

### Bài 2: Đếm số phần tử chẵn

```python
a = rd.sample(range(100), 10)
print(a)
count = 0

for i in range(len(a)):
    if a[i]%2 == 0:
        count += 1

print(count)
```

### Bài 4: Tìm số lớn nhất và nhỏ nhất trong danh sách

```python
a = rd.sample(range(100), 10)
print(a)

a_max = a[0]
a_min = a[0]

for i in range(len(a)):
    if a[i] > a_max:
        a_max = a[i]
    if a[i] < a_min:
        a_min = a[i]

print(a_max)
print(a_min)
```

### Bài 7: Ghép 2 danh sách

```python
a = [1, 1, 3, 5, 7, 9]
b = [0, 2, 2, 4, 6, 8, 10, 11]
print(a)
print(b)
# selection sort, work with every list a and b
a.extend(b)
for i in range(len(a)):
    i_min = i
    for j in range(i + 1, len(a)):
        if a[j] < a[i_min]:
            i_min = j
    a[i], a[i_min] = a[i_min], a[i]
print(a)
```

```python
a = [1, 1, 3, 5, 7, 9, 12]
b = [0, 2, 2, 4, 6, 8, 10, 11]
print(a)
print(b)
# only works with sorted lists a
for i in range(len(b)):
    for j in range(len(a)):
        if b[i] < a[j]:
            a.insert(j,b[i])
            break
        elif b[i] > a[-1]:
            a.append(b[i])
            break

print(a)
```

```python
def mergeList(a, b):

    for i in range(len(b)):
        for j in range(len(a)):
            if b[i] < a[j]:
                a.insert(j,b[i])
                break
            elif b[i] > a[-1]:
                a.append(b[i])
                break
            
    return a
```

```python
a = [1, 1, 3, 5, 7, 9, 12]
b = [0, 2, 2, 4, 6, 8, 10, 11]
c = mergeList(a, b)
print(c)
```

### Bài 8: Lọc kết quả bầu cử

```python
a = [1, 2, 3, 4, 4, 1, 1, 2, 2, 1,]
b = []

#Count number of candidates
for i in range(len(a)):
    if a[i] not in b:
        b.append(a[i])

#count votes for each candidate
for j in range(len(b)):
    count = 0
    for k in range(len(a)):
        if a[k] == b[j]:
            count += 1
    print(f'Ứng viên thứ {b[j]}: {count} phiếu')
```

```python
a = [1, 1, 3, 5, 7, 9]
b = [0, 2, 2, 4, 6, 8, 10, 11]
print(f"Initial a: {a}")
print(f"Initial b: {b}")

# Corrected logic to merge b into sorted a
for element_b in b:
    is_inserted = False
    
    # Iterate through list 'a' to find the insertion point
    for j in range(len(a)):
        if element_b < a[j]:
            # Found the spot: insert before a[j] and stop checking
            a.insert(j, element_b)
            is_inserted = True
            break
            
    # If the inner loop finished without inserting (is_inserted is still False), 
    # it means element_b is larger than all elements in 'a', so we append it.
    if not is_inserted:
        a.append(element_b)

print(f"\nCorrected Final a: {a}")
```
