```python
import random as rd

numbers = rd.sample(range(100), 10)
print(numbers)
```

```python
#sum by value
s = 0
for n in numbers:
    s += n
print(s)
```

```python
#sum by index
s = 0
for i in range(len(numbers)):
    s += numbers[i]
print(s)
```

```python
#find max
m = numbers[0]

for i in range(1, len(numbers)):
    if numbers[i] > m:
        m = numbers[i]
print(m)
```

```python
#find min
m = numbers[0]

for i in range(1, len(numbers)):
    if numbers[i] < m:
        m = numbers[i]
print(m)
```

```python
fruits = ['banana', 'mango', 'kiwi', 'apple']
prices = [100, 200, 50, 150]

max_price = max(prices)
i = prices.index(max_price)
print(fruits[i])
```

```python
print(fruits[prices.index(max(prices))])
```

```python
max_price = prices[0]
i_max = 0
for i in range(len(prices)):
    if prices[i] > max_price:
        max_price = prices[i]
        i_max = i

print(fruits[i_max])
```

```python
# find fruit that price > 100
for i in range(len(prices)):
    if prices[i] > 100:
        print(fruits[i])
```

```python
# count problem
count = 0

for fruit in fruits:
    if 'a' in fruit:
        count += 1
        print(fruit)

print(f'There are {count} fruits contains letter a')
print(fruits)
```
