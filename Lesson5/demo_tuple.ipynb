```python
a = [1, 2, 3, 4, 5] #list, element can be changed
b = (1, 2, 3, 4, 5) #tuple, element can not be changed
```

```python
print(a[0])
print(b[0])
```

```python
a[0] = 10   #list element can be changed
print(a)
```

```python
b[0] = 10   #tuple element cant be changed
```

```python
a.append(6) #list can append
print(a)
```

```python
b.append(6) #tuple cant be appended
```

```python
b = (1, 2, 3) # 1st tuple
b = b * 2       #create a new tuple
print(b)
```

```python
a = 1, 2, 3 #packing
print(a)
```

```python
m, n ,p = a #unpacking
print(m)
print(n)
print(p)
```

```python
a = 4
b = 10
#swap #1 using temp
c = a
a = b
b = c
print(a ,b)
```

```python
a = 4
b = 10
#swap #2 no temp
a = b + a
b = a - b
a = a - b
print(a ,b)
```

```python
a = 4
b = 10
#swap #3 packing and unpacking
a, b = b, a
print(a, b)
```
