```python
# dict is a collection of key-value pair
# keys are unique and immutable
# values can be access by keys
en_vn = {'apple': 'quả táo', 'banana': 'quả chuối', 'orange': 'quả cam'}
```

```python
print(en_vn['apple'])
```

```python
fruit = input('Enter a fruit in English: ')
if fruit in en_vn:
    print(f'Vietnamese fruit: {en_vn[fruit]}')
else:
    print('Sorry, I dont know')
```

```python
print(en_vn)
```

```python
# insert a new word
en_vn['mango'] = 'quả xoài'
print(en_vn)
```

```python
en_vn['durian'] = 'quả xầu riêng'
print(en_vn)
```

```python
en_vn['durian'] = 'quả sầu riêng'
print(en_vn)
```

```python
new_fruit = input('Enter a new fruit in English: ')
if new_fruit not in en_vn:
    trans = input(f'Enter the Vietnamese translation for {new_fruit}: ')
    en_vn[new_fruit] = trans
    print('Updated dictionary:', en_vn)
else:
    print(f'{new_fruit} already exists in the dictionary')
```

```python
del en_vn['orange']
print(en_vn)
```

```python
fruit = en_vn.pop('banana')
print(en_vn)
print('Remove fruit:', fruit)
```

```python
# get all keys
print(list(en_vn.keys()))
# get all values
print(list(en_vn.values()))
# get all items
print(list(en_vn.items()))
```

```python
for fruit, fruit_vn in en_vn.items():
    print(f'{fruit} => {fruit_vn}')
```
