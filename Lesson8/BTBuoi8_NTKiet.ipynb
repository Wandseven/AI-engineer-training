```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
superstore = pd.read_csv('Superstore.csv', encoding='ISO-8859-1')
```

```python
# Nhóm theo vùng, tính tổng doanh thu theo vùng
sales_by_regions = superstore.groupby('Region')['Sales'].sum()
print(sales_by_regions.head())

regions = list(sales_by_regions.index)
sales = list(sales_by_regions.values)

print(regions, sales, sep='\n')
```

```python
# Vẽ biểu đồ cột so sánh doanh thu theo vùng
plt.figure(figsize=(8,6))
plt.bar(regions, sales, color=['red', 'green', 'blue', 'purple'])
plt.title('Tổng doanh thu theo vùng')
plt.ylabel('Sales')
plt.xlabel('Region')
plt.tight_layout()
plt.show()
```

```python
# TODO: Nhóm theo Category, tính tổng doanh thu theo Category và vẽ biểu đồ cột
# TODO: Nhóm theo Category, tính tổng lợi nhuận theo Category và vẽ biểu đồ cột
# TODO: Vẽ 2 biểu đồ doanh thu và lợi nhuận ở trên cạnh nhau

#Group by category, sum sales
sales_by_category = superstore.groupby('Category')['Sales'].sum()
print(sales_by_category.head())
category = list(sales_by_category.index)
sales = list(sales_by_category.values)
#print(category, sales, sep='\n')

#Group by category, sum profit
profit_by_category = superstore.groupby('Category')['Profit'].sum()
print(profit_by_category.head())
category = list(profit_by_category.index)
profit = list(profit_by_category.values)
#print(category, profit, sep='\n')


plt.figure(figsize=(12,5))

plt.subplot(1, 2, 1)
plt.bar(category, sales, color='r')
plt.title('Tổng doanh thu theo danh mục hàng hóa')
plt.xlabel('Category')
plt.ylabel('Sales')

plt.subplot(1, 2, 2)
plt.bar(category, profit, color='b')
plt.title('Tổng lợi nhuận theo danh mục hàng hóa')
plt.xlabel('Category')
plt.ylabel('Profit')

plt.show()
```

```python
# Tính ma trận tương quan cho các cột Sales, Quantity, Discount và Profit
corr_matrix = superstore[['Sales', 'Quantity', 'Discount', 'Profit']].corr()
# TODO: Vẽ biểu đồ nhiệt đo sự tương quan
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
```

```python
# Đếm số lượng Ship Mode với value_counts
shipmode_counts = superstore['Ship Mode'].value_counts()
print(shipmode_counts)
ship_modes = list(shipmode_counts.index)
print(ship_modes)

# TODO: Vẽ biểu đồ tròn so sánh tỉ lệ các phương thức vận chuyển
plt.pie(shipmode_counts, labels=ship_modes, autopct='%1.1f%%', startangle=90)
plt.title('Tỉ lệ các phương thức vận chuyển')
```

```python
# TODO: Đếm số lượng Category, vẽ biểu đồ tròn so sánh số lượng các Category
category_count = superstore['Category'].value_counts()
print(category_count)
category = list(category_count.index)
print(category)

plt.pie(category_count, labels=category, autopct='%1.1f%%', startangle=90)
plt.title('Tỉ lệ các loại hàng hóa')
```

```python
transfers = pd.read_csv('transfers.csv')

# Xử lý dữ liệu: bỏ các dòng thiếu thông tin cần thiết
transfers = transfers.dropna(subset=['Age', 'Market_value', 'Transfer_fee'])
```

```python
vgsales = pd.read_csv('vgsales.csv')

# Loại bỏ các dòng thiếu dữ liệu cần thiết
vgsales = vgsales.dropna(subset=['Year', 'Genre', 'Global_Sales', 'NA_Sales', 'EU_Sales'])
```

```python
# TODO: Vẽ biểu đồ cột (có sắp xếp) doanh số toàn cầu theo thể loại game (Genre)

sales_by_genre = vgsales.groupby('Genre')['Global_Sales'].sum()
sales_by_genre_sorted = sales_by_genre.sort_values(ascending=True)
genre = list(sales_by_genre_sorted.index)
sales = list(sales_by_genre_sorted.values)

plt.barh(genre, sales, color='r')
plt.title('Tổng doanh số toàn cầu theo thể loại game')
plt.xlabel('Sales')
plt.ylabel('Genre')
```

```python
# TODO: Vẽ biểu đồ đường doanh số toàn cầu theo năm

sales_by_year = vgsales.groupby('Year')['Global_Sales'].sum()
year = list(sales_by_year.index)
sales = list(sales_by_year.values)

plt.plot(year, sales)
plt.title('Doanh số toàn cầu theo năm')
plt.xlabel('Year')
plt.ylabel('Global Sales')
```

```python
top3_genres = ['Action', 'Sports', 'Shooter']
top3_genres_sales = vgsales[vgsales['Genre'].isin(top3_genres)]
# TODO: Vẽ biểu đồ đường doanh số toàn cầu theo năm cho top 3 game (Action, Sports, Shooter)

action = top3_genres_sales[top3_genres_sales['Genre'].isin(['Action'])]
action_sales_by_year = action.groupby('Year')['Global_Sales'].sum()
year1 = list(action_sales_by_year.index)
action_sales = list(action_sales_by_year.values)

sport = top3_genres_sales[top3_genres_sales['Genre'].isin(['Sports'])]
sport_sales_by_year = sport.groupby('Year')['Global_Sales'].sum()
year2 = list(sport_sales_by_year.index)
sport_sales = list(sport_sales_by_year.values)

shooter = top3_genres_sales[top3_genres_sales['Genre'].isin(['Shooter'])]
shooter_sales_by_year = shooter.groupby('Year')['Global_Sales'].sum()
year3 = list(shooter_sales_by_year.index)
shooter_sales = list(shooter_sales_by_year.values)

plt.figure(figsize=(12, 5))

plt.plot(year1, action_sales, label='Action')
plt.plot(year2, sport_sales, label='Sport')
plt.plot(year3, shooter_sales, label='Shooter')
plt.legend(loc='upper right', title='Genre')
plt.title('Doanh số toàn cầu theo năm cho top 3 game (Action, Sports, Shooter)')
plt.xlabel('Year')
plt.ylabel('Global Sales')
plt.show()
```
