```python
name = input("Enter your name: ")
age = int(input("Enter your age: "))
print(f"Hello, {name}!")
print(f"You are {age} years old.")
```

```python
print(type(name), type(age))
```

```python
age = age + 5
print(f"In 5 years, you will be {age} years old.")
```

```python
print(name, age, sep='+')       # custom separator
```

```python
#format output
print(f'Name: {name}, Age: {age}')
```

```python
s = 'Jello'
a = 5
pi = 3.1415926
print(f'|{s:12}|')      # reserve 12 spaces for string
print(f'|{a:12}|')      # reserve 12 spaces for integer
print(f'|{pi:12}|')     # reserve 12 spaces for float
```

```python
print(f'|{s:>12}|')      # right align
print(f'|{a:^12}|')      # center align
print(f'|{pi:<12}|')     # left align
```
