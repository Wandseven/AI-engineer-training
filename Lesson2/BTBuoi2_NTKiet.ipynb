### 3. Phép toán cơ bản trong Python

```python
#Basic operations
a = float(input('Nhập số thứ nhất: '))
b = float(input('Nhập số thứ hai: '))
print(f'Tổng của {a} và {b} là: {a + b}')
print(f'Hiệu của {a} và {b} là: {a - b}')
print(f'Tích của {a} và {b} là: {a * b}')
if b != 0:                                  #Check if b = 0
    print(f'Thương của {a} và {b} là: {a / b}')
else:
    print('Không thể thực hiện phép chia')
```

### 4. Tính diện tích hình tròn

```python
#Area of a circle
pi = 3.14159
r1 = float(input('Bán kính hình tròn: '))
print(f'Diện tích hình tròn có bán kính {r1} là: {pi*r1**2:.2f}')    #S = pi*r^2
```

### 5. Chuyển đổi nhiệt độ

```python
# Celsius-to-Fahrenheit Converter
C = float(input('Nhiệt độ (độ C): '))
print(f'{C} độ C tương đương {C*1.8+32} độ F')       #F = C*1.8+32
```

### 6. Kiểm tra số chẵn lẻ

```python
# Check if x was odd or even
x = int(input('Nhập x: '))
if x%2 == 0:
    print(f'{x} là số chẵn')
else:
    print(f'{x} là số lẻ')
```

### 7. Hoán đổi giá trị 2 biến

```python
#Swap values of 2 variables
X = input('Nhập vào biến X: ')
Y = input('Nhập vào biến Y: ')
print(f'Giá trị của X là: {X}')
print(f'Giá trị của Y là: {Y}')
temp = Y                            #temporary variable to store Y value
Y = X
X = temp
print(f'Giá trị mới của X là: {X}')
print(f'Giá trị mới của Y là: {Y}')
```

### 8. Tính lãi suất kép

```python
P = float(input('Số tiền gốc P: '))
r = float(input('Lãi suất r: '))
n = float(input('Số lần tính lãi trong năm n:'))
t = float(input('Số năm t: '))
A = P*((1+(r/n))**(n*t))
print(f'Số tiền gốc là: {P}')
print(f'Lãi suất là: {r*100}%')
print(f'Số lần tính lãi trong 1 năm là: {n}')
print(f'Thời gian gửi là: {t} năm')
print(f'Số tiền cuối cùng là: {A:.2f}')
print(f'Số tiền thu được sau {t} năm là: {A - P:.2f}')
```
