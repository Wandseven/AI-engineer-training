### Basic

```python
class Student:
    # constructor (hàm khởi tạo), 2 tham số: name và age
    def __init__(self, name, age):
        # gán giá trị tham số cho thuộc tính đối tượng
        self.name = name
        self.age = age
    
    def hi(self):
        print(f'My name is {self.name}, I am {self.age} year old.')
```

```python
#self = john, name = 'John', age = 20
john = Student('John', 20)
print(john.name, john.age)
#call method hi() for object john
john.hi()
```

```python
paul = Student('Paul', 22)
print(paul.name, paul.age)
paul.hi()
```

### Protection levels

```python
# protection levels
paul.age = -20
print(paul.age)
```

```python
class Student:
    def __init__(self, name, age):
        self.name = name        # public attribute
        self.__age = age        # private attribute
    
    def hi(self):
        print(f'My name is {self.name}, I am {self.__age} year old.')
```

```python
john = Student('John Doe', 21)
print(john.name)        # name is accessible because it's public
```

```python
print(john.__age)       # age is not accessible because it's private
```

```python
john.hi()           # hi() method is inside the class, so it still can access to __age
```

```python
class Student:
    def __init__(self, name, age):
        self.name = name        # public attribute
        self.__age = age        # private attribute
    
    def get_age(self):          # get func to get info in class
        return self.__age
    
    def set_age(self, new_age):         # set func to change info in class
        if new_age < 18:
            print('Age can not be less than 18')
        else:
            self.__age = new_age

    def hi(self):
        print(f'My name is {self.name}, I am {self.__age} year old.')
```

```python
john = Student('John Doe', 21)
```

```python
john.set_age(22)
john.hi()
```

```python
john.set_age(17)
john.hi()
```

```python
class Employee:
    def __init__(self, name, salary):
        self.set_name(name)             # gọi luôn hàm set từ lúc khởi tạo để kiểm tra thông tin luôn
        self.set_salary(salary)
    
    def get_name(self):
        return self.__name
    
    def ask_salary(self):
        return self.__salary
    
    def set_name(self, name):
        if name == '':
            print('Name can not be empty')
        else:
            self.__name = name
    
    def set_salary(self, salary):
        if salary < 0:
            print('Salary can not be negative')
        else:
            self.__salary = salary
```

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    @property                                   #get func
    def name(self):
        return self.__name

    @name.setter                                #set func
    def name(self, new_name):
        if new_name == '':
            print('Name can not be empty')
        else:
            self.__name = new_name

    @property                                   #get func
    def salary(self):
        return self.__salary

    @salary.setter                              #set func
    def salary(self, new_salary):
        if new_salary < 0:
            print('Salary can not be negative')
        else:
            self.__salary = new_salary
```

```python
john = Employee('John', 50000)
print(john.name)
```
