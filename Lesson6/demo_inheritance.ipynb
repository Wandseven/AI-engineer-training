```python
class Person:
    def __init__(self, name):
        self.name = name
    
    @property
    def name(self):
        return self.__name
    
    @name.setter
    def name(self, name):
        if name == '':
            print('Name can not be empty')
        else:
            self.__name = name
    
    def hello(self):
        print(f'My name is {self.name}')
```

```python
p = Person('Alice')
print(p.name)
```

```python
class Student(Person):
    def __init__(self, name, grade):
        super().__init__(name)
        self.grade = grade
    
    @property
    def grade(self):
        return self.__grade
    
    @grade.setter
    def grade(self, grade):
        if grade < 0 or grade > 100:
            print('Invalde grade')
        else:
            self.__grade = grade

    # override hello method()
    def hello(self):
        super().hello()
        print(f'My grade is {self.grade}.')
```

```python
john = Student('John', 85)
print(john.name)
print(john.grade)
john.name = 'John Lennon'
print(john.name)
john.hello()
```
