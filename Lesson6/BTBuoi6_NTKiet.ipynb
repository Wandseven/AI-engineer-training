### Bài 3: Thêm phương thức vào lớp

```python
class HinhChuNhat:
    def __init__(self, dai, rong):
        self.__dai = dai
        self.__rong = rong
    
    def dien_tich(self):
        return self.__dai * self.__rong

    def chu_vi(self):
        return 2 * (self.__dai + self.__rong) 
```

```python
a = HinhChuNhat(5,4)
print(a.chu_vi())
```

### Bài 5: Kiểm tra thuộc tính riêng tư

```python
class HocSinh:
    def __init__(self,diem):
        self.diem = diem
    
    @property
    def diem(self):
        return self.__diem
    
    @diem.setter
    def diem(self, new_diem):
        if new_diem < 0 or new_diem > 10:
            print('Điểm không hợp lệ')
        else:
            self.__diem = new_diem
```

```python
hs1 = HocSinh(10)
print(hs1.diem)
hs1.diem = 11
```

### Bài 6: Kế thừa lớp

```python
class Nguoi:
    def __init__(self, name):
        self.name = name
    
    @property
    def name(self):
        return self.__name
    
    @name.setter
    def name(self, new_name):
        if new_name == '':
            print('Invalid name')
        else:
            self.__name = new_name
    
    def gioi_thieu(self):
        print(f'Xin chào, tôi là {self.__name}')
```

```python
class GiaoVien(Nguoi):
    def __init__(self,name,subject):
        super().__init__(name)
        self.subject = subject
    
    @property
    def subject(self):
        return self.__subject
    
    @subject.setter
    def subject(self, new_subject):
        if new_subject == '':
            print('Invalid Subject')
        else:
            self.__subject = new_subject
    
    def day_hoc(self):
        print(f'Giáo viên {self.name} dạy môn {self.__subject}')
```

```python
gv = GiaoVien('Thủy', 'Toán')
gv.day_hoc()
```

### Bài 7: Ghi đè phương thức

```python
class GiaoVien(Nguoi):
    def __init__(self,name,subject):
        super().__init__(name)
        self.subject = subject
    
    @property
    def subject(self):
        return self.__subject
    
    @subject.setter
    def subject(self, new_subject):
        if new_subject == '':
            print('Invalid Subject')
        else:
            self.__subject = new_subject
    
    def day_hoc(self):
        print(f'Giáo viên {self.name} dạy môn {self.__subject}')

    def gioi_thieu(self):
        print(f'Xin chào, tôi là giáo viên {self.name} dạy môn {self.__subject}')
```

```python
gv = GiaoVien('Thủy', 'Toán')
gv.gioi_thieu()
```

### Bài 8: Định nghĩa lớp PhuongTien

```python
class PhuongTien:
    def __init__(self, type):
        self.type = type
    
    @property
    def type(self):
        return self.__type
    
    @type.setter
    def type(self, new_type):
        if new_type == '':
            print('Invalid type')
        else:
            self.__type = new_type

    def di_chuyen(self):
        print(f'Loại phương tiện sử dụng là: {self.__type}')
```

```python
pt1 = PhuongTien('Xe máy')
pt1.di_chuyen()
```

### Bài 9: Tạo lớp XeMay và Oto

```python
class XeMay(PhuongTien):
    def __init__(self, type, speed, seat_count):
        super().__init__(type)
        self.speed = speed
        self.seat_count = seat_count
    
    @property
    def speed(self):
        return self.__speed
    
    @speed.setter
    def speed(self, new_speed):
        if new_speed < 0:
            print('Invalid speed')
        else:
            self.__speed = new_speed
    
    @property
    def seat_count(self):
        return self.__seat_count
    
    @seat_count.setter
    def seat_count(self, new_count):
        if new_count < 0:
            print('Invalid number of seats')
        else:
            self.__seat_count = new_count
    
    def di_chuyen(self):
        print(f'{self.type} có {self.__seat_count} ghế di chuyển với tốc độ {self.__speed} km/h')
    
```

```python
wave = XeMay('Xe máy', 30, 2)
wave.di_chuyen()
```
