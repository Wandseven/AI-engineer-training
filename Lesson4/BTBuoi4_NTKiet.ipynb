## Bài tập buổi 4

### Bài 1: Hàm tìm số lớn nhất

```python
def tim_so_lon_nhat(a, b, c):
    m = a if a > b else b
    m = c if c > m else m
    return m
```

```python
tim_so_lon_nhat(1, -15, 4)
```

### Bài 2: Hàm chuyển đổi nhiệt độ

```python
def chuyen_doi_nhiet_do(c):
    f = c*1.8 + 32
    return f
```

```python
chuyen_doi_nhiet_do(30)
```

### Bài 3: Hàm in hình chữ nhật

```python
def in_hinh_chu_nhat(n, m):
    for i in range(n):
        for j in range(m):
            print('*', end='')
        print()
```

```python
in_hinh_chu_nhat(4, 5)
```

### Bài 4: Hàm kiểm tra số chính phương

```python
def kiem_tra_so_chinh_phuong(a):
    c = a ** 0.5
    if c % 1 == 0:
        print(f'{a} là số chính phương')
    else:
        print(f'{a} không phải là số chính phương')
```

```python
kiem_tra_so_chinh_phuong(9)
```
