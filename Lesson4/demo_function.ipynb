```python
def swap(a , b):
    c = a
    a = b
    b = c
    print(f'a = {a} and b = {b} inside function')
```

```python
a = 5
b = 10
swap(a, b)
print(f'a = {a} and b = {b} outside function')
```

```python
def print_line(length):
    for _ in range(length):
        print('*', end=' ')
    print()
```

```python
n = int(input('Enter n: '))
for i in range(n):
    print_line(i+1)
```

```python
def print_line(length, c):
    for _ in range(length):
        print(c, end=' ')
    print()
```

```python
print_line(5, '?')
```

```python
def print_triangle(nline, c):
    for i in range(nline):
        print_line(i+1, c)
```

```python
print_triangle(5, '#')
```

```python
def print_line(length, c, end_line='\n'):
    for _ in range(length):
        print(c, end=' ')
    print(end_line, end=' ')
```

```python
n = 5
for i in range(n):
    print_line(n - i - 1, ' ', end_line=' ')
    print_line(2 * i + 1, '*', end_line='\n')
```

```python
# hàm tính giai thừa
def factorial(a):
    fac = 1
    for i in range(a):
        fac *= (i+1)
    return fac
```

```python
factorial(5)
```

```python
def is_prime(n):
    if n < 2:
        return False
    if n == 2:
        return True
    
    for i in range(2, n):
        if n % i == 0:
            return False
    return True
```

```python
n = int(input('Enter n: '))
if is_prime(n):
    print(f'{n} is prime number')
else:
    print(f'{n} is not prime number')
```

```python
def print_primes(n):
    for i in range(2, n + 1):
        if is_prime(i):
            print(i, end=' ')
    print()
```

```python
print_primes(15)
```
